Question 4_2

```python
import numpy as np

# Given probabilities
probabilities = [
    [0.15, 0.09, 0.06],
    [0.15, 0.09, 0.06],
    [0.20, 0.12, 0.08]
]

# Flatten the probability matrix
flat_probabilities = [p for row in probabilities for p in row]

# Generate 300 pseudo-random numbers
random_numbers = np.random.rand(300)

# Classify the observations based on the specified conditions
classified_observations = np.zeros(300, dtype=int)

for i in range(1, 10):
    if i == 1:
        condition = random_numbers < flat_probabilities[0]
    else:
        condition = np.logical_and(
            random_numbers >= np.sum(flat_probabilities[:i-1]),
            random_numbers < np.sum(flat_probabilities[:i])
        )
    classified_observations[condition] = i

# Display the counts in each cell
print("Counts in each cell:")
for i in range(1, 10):
    count = np.sum(classified_observations == i)
    print(f"Cell {i}: {count} observations")

# Example of the first few observations
print("\nExample of the first few observations:")
print(classified_observations[:10])
```

Question 4_4

```python
import numpy as np
from scipy.stats import chi2_contingency

# Given probabilities
probabilities = [
    [0.15, 0.09, 0.06],
    [0.15, 0.09, 0.06],
    [0.20, 0.12, 0.08]
]

# Observed counts in each cell (use the previously generated data)
observed_counts = np.zeros((3, 3), dtype=int)

for i in range(1, 10):
    row, col = divmod(i - 1, 3)
    observed_counts[row, col] = np.sum(classified_observations == i)

# Expected counts based on probabilities
total_observations = 300
expected_counts = np.array(probabilities) * total_observations

# Perform the chi-squared goodness-of-fit test
chi2_stat, p_value, _, expected = chi2_contingency(observed_counts)

# Display the results
print(f"Chi-squared statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print("Expected counts based on probabilities:")
print(expected)

# Test the hypothesis at a significance level of 0.05
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis. The values form a random sample from the χ² distribution.")
else:
    print("\nFail to reject the null hypothesis. The values may form a random sample from the χ² distribution.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom_test

# Set the sample size
n = 100

# Significance level (alpha)
alpha = 0.05

# Values of p for the alternative hypothesis
p_values = np.linspace(0, 1, 100)

# Calculate the power for each p
power_values = [1 - binom_test(0.5 * n, n, p, alternative='two-sided') for p in p_values]

# Plot the power diagram
plt.figure(figsize=(8, 6))
plt.plot(p_values, power_values, label='Power Curve')
plt.axhline(y=alpha, color='r', linestyle='--', label='Significance Level (alpha)')
plt.title('Power Diagram for Binomial Test')
plt.xlabel('Probability (p)')
plt.ylabel('Power')
plt.legend()
plt.show()
```

As can be seen, power has the highest value when p is farthest from 0.5 and the lowest value when p is close to 0.5. p is far from the null hypothesis value of 0.5, which indicates the ability of the test to correctly reject the null hypothesis if it is false. A power plot shows how the power of the test changes when the p value of the opposing hypothesis changes. Power is usually higher when p is far from the null hypothesis value of 0.5, which indicates the ability of the test to correctly reject the null hypothesis when it is false.

Qestion 20_1,2,3

```python
import numpy as np
from scipy.stats import beta, wilcoxon

# Set parameters
a = 2
b = 5
sample_size = 50
alpha = 0.05

# Part 1: Hypothesis Test with H0: M = 0.4, H1: M > 0.4
np.random.seed(42)
sample = beta.rvs(a, b, size=sample_size)
M = np.median(sample)
statistic, p_value = wilcoxon(sample - 0.4, alternative='greater')

print(f"1) Median of the sample: {M}")
print(f"   Test Statistic: {statistic}")
print(f"   P-value: {p_value}")

if p_value < alpha:
    print("   Reject H0: There is evidence to suggest that M > 0.4.")
else:
    print("   Fail to reject H0: There is not enough evidence to suggest that M > 0.4.")

# Part 2: Calculate Power for H0: M = 0.4, H1: M > 0.4
alternative_median = 0.6
sample_alt = beta.rvs(a, b, size=sample_size) + (alternative_median - 0.4)
statistic_alt, p_value_alt = wilcoxon(sample_alt - 0.4, alternative='greater')
power = 1 - p_value_alt

print(f"\n2) Power of the test: {power}")

# Part 3: Increase Hypothesized Median to 0.6
statistic_06, p_value_06 = wilcoxon(sample - 0.6, alternative='greater')

print(f"\n3) Test Statistic: {statistic_06}")
print(f"   P-value: {p_value_06}")

if p_value_06 < alpha:
    print("   Reject H0: There is evidence to suggest that M > 0.6.")
else:
    print("   Fail to reject H0: There is not enough evidence to suggest that M > 0.6.")

# Calculate power for H0: M = 0.6, H1: M > 0.6
power_06 = 1 - p_value_06
print(f"   Power of the test: {power_06}")
```

Question 21_1,2

```python
import numpy as np
from scipy.stats import f

np.random.seed(42)

# Part 2: Generate samples from two normal distributions
sample_size = 100
sigma_x = np.sqrt(3)
sigma_y = np.sqrt(10)

X_samples = np.random.normal(0, sigma_x, size=sample_size)

Y_samples = np.random.normal(0, sigma_y, size=sample_size)

# Part 1: F-test for equality of variances
# H0: σ_x = σ_y
# Ha: σ_x ≠ σ_y

# Calculate the sample variances
var_x = np.var(X_samples, ddof=1)  # ddof=1 for unbiased estimate
var_y = np.var(Y_samples, ddof=1)

# Calculate the F-statistic
F_statistic = var_x / var_y

# Degrees of freedom for the F-distribution
df_x = sample_size - 1
df_y = sample_size - 1

# Critical value for a two-sided test with alpha=0.05
alpha = 0.05
critical_value = f.ppf(1 - alpha / 2, df_x, df_y)

# Compare the F-statistic to the critical value
if F_statistic > critical_value or F_statistic < 1 / critical_value:
    print(f"Reject H0: Variances are not equal")
else:
    print("Fail to reject H0: Variances are equal")

print("\nF-statistic:", F_statistic)
print("Critical Value:", critical_value)
```

Question 22_a,b,c

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom


number_of_sons = np.arange(12)
frequency = np.array([7, 45, 181, 478, 829, 1112, 1343, 1033, 670, 286, 104, 24])

# a) Introduce statistics for hypothesis testing
n = 12  # Number of trials (children in a family)
p = 0.5  # Probability of having a son
expected_frequency = binom.pmf(number_of_sons, n, p) * np.sum(frequency)

print(expected_frequency.shape)
print(frequency.shape)

# Chi-square statistic calculation
chi_square_statistic = np.sum((frequency[:-1] - expected_frequency[:-1]) ** 2 / expected_frequency[:-1])

# b) Simulation and histogram
num_simulations = 10000
simulated_data = np.random.binomial(n, p, size=(num_simulations, len(number_of_sons)))
simulated_frequencies = np.apply_along_axis(lambda x: np.histogram(x, bins=number_of_sons)[0], axis=1, arr=simulated_data)

# c) Calculate p-value
expected_frequency_broadcasted = np.tile(expected_frequency[:-1], (num_simulations, 1))
simulated_chi_square_statistics = np.sum((simulated_frequencies - expected_frequency_broadcasted) ** 2 / expected_frequency_broadcasted, axis=1)
p_value = np.mean(simulated_chi_square_statistics >= chi_square_statistic)

plt.hist(simulated_chi_square_statistics, bins=50, density=True, alpha=0.7, color='blue', label='Simulated Distribution')
plt.axvline(x=chi_square_statistic, color='red', linestyle='dashed', linewidth=2, label='Observed Statistic')
plt.title('Distribution of Chi-square Statistic')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Density')
plt.legend()
plt.show()

print(f"Chi-square Statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")
print(f"Is H0 rejected? {'Yes' if p_value < 0.05 else 'No'}")
```
